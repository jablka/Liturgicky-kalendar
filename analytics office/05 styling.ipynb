```python
def styling_factory(styler):
    styler.set_uuid('LC') # setting UUID, otherwise would be set to random string.
    styler.hide(level='moz') # skrývame nepotrebné levely v indexe,
    styler.set_table_styles([
        # {'selector': 'table', 'props': 'width: 75%; margin-left: auto; margin-right: auto; font-family: Arial, Helvetica, sans-serif; line-height: 1.5;'},
        {'selector':'th,td', 'props':'border: 1px solid black; border-collapse: collapse; padding: 1rem;'},
        {'selector': 'td', 'props': 'white-space: pre-wrap;'},
        {'selector': 'th', 'props': 'text-wrap: nowrap; font-weight: normal;'}, 
        {'selector': '.col_heading,.index_name', 'props': 'font-weight: bold;'}, # vieme kombinovať selectory
        ])
    
    ts = ( 'style="width: 75%; margin-left: auto; margin-right: auto;'
            'font-family: Arial, Helvetica, sans-serif; line-height: 1.5;'
            'border: 1px solid black; border-collapse: collapse; padding: 1rem;"' ) # single string
    styler.set_table_attributes(ts)

    return styler

df.style.pipe(styling_factory).to_html('table.html', doctype_html = True)
```
