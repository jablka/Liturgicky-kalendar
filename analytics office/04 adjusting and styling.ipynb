```python
import pandas as pd
```

```python
df = pd.read_pickle('second_dataframe.pickle')
df
```

```python
# vytvoríme si samostatný frame, pozostávajúci z df.index
# ten si postupne upravíme (pridáme stĺpce / zmažeme) do požadovanej formy.
# nakoniec vytvoríme z framu Multiindex, a napojíme ho do pôvodného df.index
```

```python
# vytvoríme si samostatný frame, pozostávajúci z df.index
frame_index = df.index.to_frame().reset_index(drop=True)
frame_index
```

```python
# konverzia datum string na datetime
frame_index['datum'] = pd.to_datetime(frame_index['datum']) 
frame_index.head()
```

```python
# generujeme nový stĺpec: deň v mesiaci
d_mes = frame_index['datum'].dt.day.astype(str) + '. ' + frame_index['datum'].dt.month_name('sk_SK').str.lower()
frame_index.insert(1, 'd_mes', d_mes)
frame_index.head()
```

```python
# generujeme nový stĺpec: deň v týždni
d_tyz = frame_index['datum'].dt.day_name('sk_SK').str.lower()
frame_index.insert(2, 'd_tyz', d_tyz)
frame_index.head()
```

```python
# datum back to string
frame_index['datum'] = frame_index.loc[:,'datum'].astype(str)
frame_index.head()
```

```python
# generujeme stĺpec deň - ako bude vyzerať vo výslednej html tabuľke.
den = frame_index.d_mes + ', ' + frame_index.d_tyz + '<br><i>' + frame_index.meniny +'</i>'
frame_index.insert(1, 'den', den)
frame_index.head()
```

```python
# mažeme prebytočné stĺpce
frame_index.drop(['d_tyz','d_mes','meniny'], axis='columns', inplace=True)
frame_index.head()
```

```python
# vytvárame MultiIndex
new_multiindex = pd.MultiIndex.from_frame(frame_index)
new_multiindex[:5]
```

```python
# prepisujeme df.index novým multiindexom
df.index = new_multiindex # alebo: df = df.set_index(new_multiindex)
df
```

```python
# dopĺňam diakritiku do názvov
df.index.names = ['dátum', 'deň', 'moz']
df.head()
```

```python
# skrývame nepotrebné levely v indexe,
# a ukladáme do stringu
s = df.style.hide(level=['dátum','moz']).to_html()
```

```python
# css štýly tabuľky
css = """
<style>
  table {
    width: 75%;
    margin-left: auto;
    margin-right: auto;
    font-family: Arial, Helvetica, sans-serif;
    line-height: 1.5;
  }

  table,
  th,
  td {
    border: 1px solid black;
    border-collapse: collapse;
    padding: 1rem;
  }

  td {
    white-space: pre-wrap;
  }

  th {
    text-wrap: nowrap;
    font-weight: normal;
  }

  .col_heading,
  .index_name {
    font-weight: bold;
  }

  .col1 {
    /* font-size: smaller; */
  }
 
</style>
"""
```

```python
# tvorba výsledného html súboru.
with open('resulting_table.html', mode='w', encoding='utf8') as f:
    f.write(css)
    f.write(s)
```
